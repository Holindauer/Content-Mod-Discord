# Data Preperation for Test Bot
The test bot will send multiple toxic messages to the discord server. This data preperation seperates out toxic columns for this purpose.

```python
import pandas as pd

# Read in the data
df = pd.read_csv('..//..//data//train.csv')
df.head()
```

```python
#drop "id" column
df.drop("id", axis=1, inplace=True)
df.head()
```

```python
df.shape
```

```python
toxic_cols = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]

#for each row, check if any toxic_col is one. If True, then add to the toxic_df
toxic_df = df[df[toxic_cols].sum(axis=1) > 0]
toxic_df.head()
```

```python
#drop toxic cols
toxic_df = toxic_df.drop(toxic_cols, axis=1)
toxic_df.head()
```

```python
#save toxic_df as csv
toxic_df.to_csv('toxic_df.csv', index=False)
```
